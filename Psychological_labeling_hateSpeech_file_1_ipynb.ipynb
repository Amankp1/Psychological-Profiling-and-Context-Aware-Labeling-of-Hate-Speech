```python
import evaluate
metric = evaluate.load("accuracy")
```

```python
# Define label mapping
label_mapping = {"Depression": 0, "Stress": 1, "Loneliness": 2}
reverse_mapping = {0: "Depression", 1: "Stress", 2: "Loneliness"}

# Convert string labels to numerical values
numeric_labels = [label_mapping[label] for label in test_labels]

correct = 0  # Counter for correct predictions
total = len(test_texts)

# Evaluate the model
for text, true_label in zip(test_texts, numeric_labels):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
        predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)

    predicted_class = torch.argmax(predictions).item()
    predicted_label = reverse_mapping[predicted_class]
    true_label_text = reverse_mapping[true_label]

    # Add prediction to metric
    metric.add_batch(predictions=[predicted_class], references=[true_label])

    # Check if the prediction is correct
    if predicted_class == true_label:
        correct += 1
        status = "✅ Correct"
    else:
        status = "❌ Wrong"

    # Print result for each text
    print(f"Text: {text}")
    print(f"True Label: {true_label_text} | Predicted: {predicted_label} → {status}")
    print("-" * 80)

# Compute overall accuracy
accuracy = correct / total * 100
print(f"\nOverall Accuracy: {accuracy:.2f}%")
print(metric.compute())
```
